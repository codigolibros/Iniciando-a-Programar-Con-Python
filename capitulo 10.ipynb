### Creación de objetos persona de tipo lista 

```python
# Creación de objetos persona de tipo lista 
personas = []
# Creación del objeto Persona con valores específicos
personas.append(Persona("123456789", "Juan", "Pérez"))  
# Creación del objeto Persona con valores específicos
personas.append(Persona("987654321", "María", "Gómez"))  
# Creación del objeto Persona con valores específicos
personas.append(Persona("456789123", "Carlos", "Sánchez"))  

# Mostrar información de cada persona en la lista
for persona in personas:
    # Llamada al método mostrarInformacion() de cada objeto Persona
    persona.mostrarInformacion()  
```

### Ordenar la lista de personas por documento

```python
# Ordenar la lista de personas por documento utilizando el método sort()
personas.sort(key=lambda persona: persona.documento)

# Mostrar información de cada persona en la lista ordenada
for persona in personas:
    persona.mostrarInformacion()
```

### Ordenar la lista de personas por apellidos

```python
# Ordenar la lista de personas por apellidos utilizando el método sort()
personas.sort(key=lambda persona: persona.apellidos)

# Mostrar información de cada persona en la lista ordenada
for persona in personas:
    persona.mostrarInformacion()
```

### Insertar el nuevo objeto Persona

```python
# Crear un nuevo objeto Persona
nueva_persona = Persona("111111111", "Luis", "González")

# Insertar el nuevo objeto Persona en la segunda posición de la lista
personas.insert(1, nueva_persona)

# Mostrar información de cada persona en la lista actualizada
for persona in personas:
    persona.mostrarInformacion()
```

### Eliminar y retornar elementos de una Lista

## Conjuntos

```python
# Declaración de un conjunto con cinco ciudades de Colombia
ciudadesColombia = {"Bogotá", "Medellín", "Cali", "Barranquilla", "Cartagena"}

# Imprimir el conjunto de ciudades
print(ciudadesColombia)
```

### Union

```python
# Conjunto ciudadesColombia
ciudadesColombia = {"Bogotá", "Medellín", "Cali", "Barranquilla", "Cartagena"}

# Conjunto otroConjuntoCiudades
otroConjuntoCiudades = {"Cali", "Bucaramanga", "Santa Marta"}

# Realizar la unión de los conjuntos
unionCiudades = ciudadesColombia.union(otroConjuntoCiudades)

# Imprimir el resultado de la unión
print(unionCiudades)
```

### Intersección

```python
# Conjunto ciudadesColombia
ciudadesColombia = {"Bogotá", "Medellín", "Cali", "Barranquilla", "Cartagena"}

# Conjunto otroConjuntoCiudades
otroConjuntoCiudades = {"Cali", "Bucaramanga", "Santa Marta"}

# Realizar la intersección de los conjuntos
interseccionCiudades = ciudadesColombia.intersection(otroConjuntoCiudades)

# Imprimir el resultado de la intersección
print(interseccionCiudades)
```

### diferencia de conjuntos

```python
# Conjunto ciudadesColombia
ciudadesColombia = {"Bogotá", "Medellín", "Cali", "Barranquilla", "Cartagena"}

# Conjunto otroConjuntoCiudades
otroConjuntoCiudades = {"Cali", "Bucaramanga", "Santa Marta"}

# Calcular la diferencia de conjuntos
diferenciaCiudades = ciudadesColombia.difference(otroConjuntoCiudades)

# Imprimir el resultado de la diferencia
print(diferenciaCiudades)
```

### Diferencia simétrica de conjuntos

```python
# Conjunto ciudadesColombia
ciudadesColombia = {"Bogotá", "Medellín", "Cali", "Barranquilla", "Cartagena"}

# Conjunto otroConjuntoCiudades
otroConjuntoCiudades = {"Cali", "Bucaramanga", "Santa Marta"}

# Calcular la diferencia simétrica de conjuntos
diferenciaSimetricaCiudades = ciudadesColombia.symmetric_difference(otroConjuntoCiudades)

# Imprimir el resultado de la diferencia simétrica
print(diferenciaSimetricaCiudades)
```

### Operaciones básicas

```python
# Operaciones básicas
# Crear un conjunto
conjunto1 = {1, 2, 3, 4, 5}

# Copiar el conjunto
conjunto2 = conjunto1.copy()
print(conjunto1)  # {1, 2, 3, 4, 5}
print(conjunto2)  # {1, 2, 3, 4, 5}

# Verificar si dos conjuntos tienen intersección nula
conjunto3 = {6, 7, 8}
print(conjunto1.isdisjoint(conjunto3))  # True

# Verificar si un conjunto es subconjunto de otro
conjunto4 = {1, 2, 3}
print(conjunto4.issubset(conjunto1))  # True

# Verificar si un conjunto es superconjunto de otro
print(conjunto1.issuperset(conjunto4))  # True

# Obtener una lista ordenada a partir del conjunto
conjunto_ordenado = sorted(conjunto1)
print(conjunto_ordenado)  # [1, 2, 3, 4, 5]

# Obtener la suma de todos los elementos del conjunto
suma_conjunto = sum(conjunto1)
print(suma_conjunto)  # 15

# Obtener la longitud del conjunto
print(len(conjunto1))  # 5

# Obtener el elemento más pequeño del conjunto
print(min(conjunto1))  # 1

# Obtener el elemento más grande del conjunto
print(max(conjunto1))  # 5
```

```python
# Declaración de una tupla de meses del año
meses = ("enero", "febrero", "marzo", "abril", "mayo", "junio", "julio", "agosto", "septiembre", "octubre", "noviembre", "diciembre")

# Obtener el número de elementos en la tupla
print(len(meses))  # Resultado: 12

# Obtener el índice de un elemento en la tupla
print(meses.index("abril"))  # Resultado: 3

# Contar el número de veces que aparece un elemento en la tupla
print(meses.count("marzo"))  # Resultado: 1

# Convertir la tupla en una lista
meses_lista = list(meses)
print(meses_lista)  # Resultado: ['enero', 'febrero', 'marzo', 'abril', 'mayo', 'junio', 'julio', 'agosto', 'septiembre', 'octubre', 'noviembre', 'diciembre']

# Convertir la lista nuevamente en una tupla
meses_tupla = tuple(meses_lista)
print(meses_tupla)  # Resultado: ('enero', 'febrero', 'marzo', 'abril', 'mayo', 'junio', 'julio', 'agosto', 'septiembre', 'octubre', 'noviembre', 'diciembre')

# Comprobar si un elemento está presente en la tupla
print("julio" in meses)  # Resultado: True

# Obtener una sub-tupla de la tupla original
sub_tupla = meses[2:6]
print(sub_tupla)  # Resultado: ('marzo', 'abril', 'mayo', 'junio')
```

### Diccionario inventario

```python
# Diccionario inventario
inventario = {
    # primer articulo
    "12345": {
        "nombre": "Camisa",
        "cantidad": 50,
        "precioUnitario": 25000.00
    },
    # segundo articulo
    "67890": {
        "nombre": "Pantalón",
        "cantidad": 30,
        "precioUnitario": 39000.00
    },
    # tercer articulo
    "54321": {
        "nombre": "Zapatos",
        "cantidad": 20,
        "precioUnitario": 59000.00
    }
}

# Acceso a los elementos del inventario
print(inventario["12345"])  # Acceder a un artículo específico por su código
print(inventario["67890"]["nombre"])  # Acceder al nombre de un artículo específico
print(inventario["54321"]["cantidad"])  # Acceder a la cantidad de un artículo específico
```

```python
# Acceder a los detalles del artículo con código "12345"
articulo12345 = inventario["12345"]
print(articulo12345)
```

```python
# Acceder al nombre del artículo con código "12345"
nombreArticulo = inventario["12345"]["nombre"]
print(nombreArticulo)
```

###  Buscar artículo

```python
# Nombre del artículo a buscar
nombreArticulo = "Camisa"

# Variable para controlar si se encuentra el artículo
encontrado = False

# Búsqueda del artículo por nombre
for codigo, articulo in inventario.items():
    if articulo["nombre"] == nombreArticulo:
        print("Código:", codigo)
        print("Nombre:", articulo["nombre"])
        print("Cantidad:", articulo["cantidad"])
        print("Precio unitario:", articulo["precioUnitario"])
        encontrado = True
        break

# Si no se encuentra el artículo, imprimir mensaje
if not encontrado:
    print("El artículo", nombreArticulo, "no se encuentra en el inventario.")
```
