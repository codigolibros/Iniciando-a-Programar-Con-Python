## Identificadores de Python
Como sugiere el nombre, los identificadores nos ayudan a diferenciar una entidad de otra. Las entidades de Python, como clases, funciones y variables, se llaman identificadores.

• Puede ser una combinación de letras mayúsculas o minúsculas (de a a z o de A a Z).
• Puede ser cualquier dígito (0 a 9) o un guión bajo (_).
• Se deben seguir las reglas generales para escribir identificadores en Python.
• No puede comenzar con un dígito. Por ejemplo, 1variable no es válido, mientras que variable1 es válido.

| Identificadores válidos | Ejemplos                               |
|------------------------|----------------------------------------|
| Letras mayúsculas      | Nombre, MiClase, HTTPError             |
| Letras minúsculas      | edad, mi_variable, calcular_area       |
| Dígitos                | variable1, suma2, valor3               |
| Guiones bajos          | mi_variable, mi_funcion, mi_modulo     |
| Inicio con letra       | nombre, edad, mi_variable              |
| Inicio con guión bajo  | _variable, _mi_funcion, _mi_modulo     |

## Suites

```python
# Ejemplo de indentación con Suite def

def saludar(nombre):
    print("Hola, " + nombre)
    print("¡Bienvenido!")
    print("Es un gusto tenerte aquí.")

# Llamada a la función
saludar("Juan")
```

```python
# Ejemplo de indentación con Suite if

def evaluar_edad(edad):
    if edad >= 18:
        print("Es mayor de edad.")
        print("Puede ingresar al establecimiento.")
    else:
        print("Es menor de edad.")
        print("No puede ingresar al establecimiento.")

# Llamada a la función
evaluar_edad(20)
```

```python
# Ejemplo de indentación con Suite de un bucle for
for i in range(5):
    print(i)
    print("Dentro del bucle")
print("Fuera del bucle")
```

```python
# Ejemplo comentarios de una sola línea

# Puede usar comentarios para explicar el 
# propósito de su código
# o para documentar partes específicas del mismo.

# También Puede utilizar comentarios para desactivar 
#temporalmente
# una porción de código que no deseas ejecutar.

# Declaración de variables
nombre = "Juan"  # Nombre del usuario
edad = 25  # Edad del usuario

# Cálculo del año de nacimiento
anioActual = 2023
anioNacimiento = anioActual - edad

# Imprimir el resultado
print("El año de nacimiento de", nombre, "es:", 
      anioNacimiento)
```

## Sentencias multilínea

```python
# Declaración multilínea con paréntesis y \
total = (10 +
         20 +
         30)
print("El resultado de la suma es:", total)
# Declaración multilínea con corchetes
lista = [
    'elemento 1',
    'elemento 2',
    'elemento 3'
]
# Comando multilínea con \
print("La lista contiene los siguientes \
elementos:")
for elemento in lista:
    print(elemento)
# Declaración multilínea con llaves
diccionario = {
    'clave1': 'valor1',
    'clave2': 'valor2',
    'clave3': 'valor3'
}
# Comando multilínea con \
print("El diccionario tiene las siguientes claves y \
valores:")
for clave, valor in diccionario.items():
    print(clave, ":", valor)
```
