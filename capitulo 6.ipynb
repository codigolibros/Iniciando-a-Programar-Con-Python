# Operadores

los operadores son símbolos especiales que se utilizan para realizar operaciones en variables, constantes y valores. Los operadores nos permiten realizar cálculos matemáticos, comparar valores, combinar valores booleanos y manipular estructuras de datos

## Operadores básicos

En Python, los operadores son los símbolos especiales que pueden manipular el valor de los operandos. Por ejemplo, consideremos la expresión 1 + 2 = 3. Aquí, 1 y 2 se llaman operandos, que son los valores en los que los operadores operan y el símbolo + se llama operador. El lenguaje Python admite los siguientes tipos de operadores:

• Operadores de asignación

• Operadores aritméticos

• Operadores de comparación o relación

• Operadores lógicos

• Operadores de pertenencia

• Operadores de identidad

• Operadores a nivel de bits

## Operadores de asignación 

El operador de asignación básico "=" se utiliza para asignar un valor a una variable. Otros operadores de asignación, como "+=", "-=", "*=", "/=", etc., combinan una operación matemática con la asignación. Por ejemplo, "a += 5" es equivalente a "a = a + 5", lo que significa que se suma 5 al valor actual de "a" y se asigna el resultado nuevamente a "a".. Ver Tabla 1-5 y Listado 1-9.

Estos operadores de asignación son útiles para simplificar y abreviar el código cuando se realiza una operación matemática y se asigna el resultado a la misma variable.

Recuerda que los operadores de asignación en Python pueden ser utilizados con diferentes tipos de datos y no solo con números. Puedes aplicarlos a variables de diferentes tipos, como enteros, flotantes, cadenas, listas, etc.

Listado 1-9. Ejemplo de código para operadores de asignación.


```python
# Ejemplo de Operadores de Asignación

# Asignación básica
a = 10
print("El valor de a es:", a)

# Asignación de suma
a += 5
print("Después de a += 5, el valor de a es:", a)

# Asignación de resta
a -= 3
print("Después de a -= 3, el valor de a es:", a)

# Asignación de multiplicación
a *= 2
print("Después de a *= 2, el valor de a es:", a)

# Asignación de división
a /= 4
print("Después de a /= 4, el valor de a es:", a)

# Asignación de módulo
a %= 2
print("Después de a %= 2, el valor de a es:", a)

# Asignación de potencia
a **= 3
print("Después de a **= 3, el valor de a es:", a)

# Asignación de división entera
a //= 5
print("Después de a //= 5, el valor de a es:", a)
```

## Operadores aritméticos

```python
# Ejemplo de Operadores Aritméticos
a = 10
b = 5

# Suma
resultadoSuma = a + b
print("La suma de a y b es:", resultadoSuma)

# Resta
resultadoResta = a - b
print("La resta de a y b es:", resultadoResta)

# Multiplicación
resultadoMultiplicacion = a * b
print("La multiplicación de a y b es:", resultadoMultiplicacion)

# División
resultadoDivision = a / b
print("La división de a y b es:", resultadoDivision)

# Módulo
resultadoModulo = a % b
print("El módulo de a y b es:", resultadoModulo)

# Potencia
resultadoPotencia = a**b
print("El resultado de elevar a a la potencia b es:", resultadoPotencia)

# División entera
resultadoDivisionEntera = a // b
print("La división entera de a y b es:", resultadoDivisionEntera )
```

## Operadores de comparación o relacionales

```python
# Ejemplo de Operadores de Comparación o Relacionales
a = 10
b = 5

# Igual a
igualA = a == b
print("¿a es igual a b?", igualA)

# Diferente de
diferenteDe = a != b
print("¿a es diferente de b?", diferenteDe)

# Menor que
menorQue = a < b
print("¿a es menor que b?", menorQue)

# Mayor que
mayorQue = a > b
print("¿a es mayor que b?", mayorQue)

# Menor o igual que
menorIgualQue = a <= b
print("¿a es menor o igual que b?", menorIgualQue)

# Mayor o igual que
mayorIgualQue = a >= b
print("¿a es mayor o igual que b?", mayorIgualQue)
```

## Operadores lógicos

```python
# Ejemplo de operadores lógicos
a = 5
b = 10
c = 3

# Operador lógico AND
if a > 0 and b < 20:
    print("Ambas condiciones son verdaderas")

# Operador lógico OR
if a < 0 or c > 0:
    print("Al menos una de las condiciones es verdadera")

# Operador lógico NOT
if not a == b:
    print("La condición es falsa")

# Uso de operadores lógicos con estructuras de control
if (a > 0 and b > 0) or c == 3:
    print("Se cumple una de las condiciones")

# Combinación de operadores lógicos
if (a > 0 or b > 0) and c == 3:
    print("Se cumplen todas las condiciones")

```

## Operadores de Membresía

```python
# Ejemplo  operadores de membresía
lista = [1, 2, 3, 4, 5]

# Comprobar si un valor está presente en la lista
if 3 in lista:
    print("El valor 3 está en la lista")

# Comprobar si un valor no está presente en la lista
if 6 not in lista:
    print("El valor 6 no está en la lista")
```

## Operadores de identidad

```python
# Ejemplo operadores de identidad
x = [1, 2, 3]
y = [1, 2, 3]
z = x

# Comparar la identidad de dos objetos
print(x is y)       # Salida: False
print(x is z)       # Salida: True

# Comparar la no identidad de dos objetos
print(x is not y)   # Salida: True
print(x is not z)   # Salida: False
```

## Operadores bit a bit

```python
# Ejemplo 0peradores bit a bit
a = 5   # Representación binaria: 0101
b = 3   # Representación binaria: 0011

# AND a nivel de bits
resultado_and = a & b
print("Resultado de a & b:", resultado_and)  # Salida: 1 (binario: 0001)

# OR a nivel de bits
resultado_or = a | b
print("Resultado de a | b:", resultado_or)  # Salida: 7 (binario: 0111)

# XOR a nivel de bits
resultado_xor = a ^ b
print("Resultado de a ^ b:", resultado_xor)  # Salida: 6 (binario: 0110)

# Complemento a nivel de bits
complemento_a = ~a
print("Complemento de a:", complemento_a)  # Salida: -6

# Desplazamiento a la izquierda
desplazamiento_izquierda = a << 2
print("Desplazamiento a la izquierda de a:", desplazamiento_izquierda)  # Salida: 20

# Desplazamiento a la derecha
desplazamiento_derecha = a >> 1
print("Desplazamiento a la derecha de a:", desplazamiento_derecha)  # Salida: 2
```
